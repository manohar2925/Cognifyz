## Task: Price Range Distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
filepath="D:\\manu\\data_science\\cognifyz\\Datasets.csv"
df=pd.read_csv(filepath)
```

```python
df.head()
```

```python
df.columns
```

```python
price_range_percentage = (df['Price range'].value_counts() / len(df)) * 100
print("Percentage of Restaurants in Each Price Range:")
print(price_range_percentage)
```

```python
plt.figure(figsize=(8, 6))
ax=sns.countplot(x='Price range', data=df, palette='Set2')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
filepath="D:\\manu\\data_science\\cognifyz\\Datasets.csv"
df=pd.read_csv(filepath)
```

```python
df.columns
```

```python
city=df['City'].unique()
city
plt.figure(figsize=(20,6))
sns.countplot(x='City', data=df, palette='viridis')
plt.title('Number of Restaurants in Each City')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
restaurants_count=df['City'].value_counts()
restaurants_count
```

```python
# City with highest restaruants

highest_restaurants=df['City'].value_counts().idxmax()
highest_restaurants
```

```python
# The average rating for restaurants in each city

average_rating=df.groupby('City')['Aggregate rating'].mean()
#print(len(average_rating))
print(average_rating)
```

```python
# the city with the highest average rating.

city_order=average_rating.sort_values(ascending=False)
top_city=city_order.head(1)
print('city with the highest number of restaurants:',highest_restaurants)
#print('average rating forrestaurants in each city:',average_rating)
print('city with the highest average rating:',top_city)
```
